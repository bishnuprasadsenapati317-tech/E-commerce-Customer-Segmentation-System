## Data Preprocessing

### 1. Handle Missing values

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

### Feature Engineering

```python
#Age 
df["Age"] = 2026 - df["Year_Birth"] # can use df["Year_Birth"].max()
# print( df["Year_Birth"].max())
```

```python
# Customer Joining date
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)

reference_date = df["Dt_Customer"].max()

df["customer_tenure_days"] = (reference_date - df["Dt_Customer"]).dt.days
```

```python
#Spending 
df["Total_Spending"] = df["MntWines"] + df["MntGoldProds"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"]
```

```python
#Total children
df["Total_Children"] = df["Kidhome"] + df["Teenhome"]
df.drop("total_spending", axis=1)
```

```python
#Education

df["Education"].value_counts()

df["Education"] = df["Education"].replace({
    "2n Cycle": "Undergraduate","Basic": "Undergraduate",
     "Graduation": "Graduate", 
    "Master": "Postgraduate", "PhD": "Postgraduate"
})
```

```python
df["Education"].value_counts()
```

```python
#Marital Status
df["Marital_Status"].value_counts()

df["Living_With"] = df["Marital_Status"].replace({
    "Married": "Partner","Together": "Partner",
     "Single": "Alone","Divorced": "Alone",
    "Absurd": "Alone", "YOLO": "Alone",
    "Widow":"Alone"
})
```

```python
df["Living_With"].value_counts()
```

```python
df.columns
```

```python
cols = ["ID", "Marital_Status", "Kidhome", "Teenhome","total_spending", "Year_Birth", "Dt_Customer" ]
spending_cols = ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]

cols_to_drop = cols + spending_cols
df_cleaned = df.drop(columns=cols_to_drop)
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
cat_cols = ["Education", "Living_With"]
enc_cols = ohe.fit_transform(df_cleaned[cat_cols])
```

```python
enc_df = pd.DataFrame(enc_cols.toarray(), columns=ohe.get_feature_names_out(cat_cols), index=df_cleaned.index)
```

```python
enc_df.head()
```

```python
df_encoded = pd.concat([df_cleaned.drop(columns=cat_cols),enc_df], axis=1)
```

```python
df_encoded.shape
```

```python
df_encoded.head()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

X = df_encoded
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

# Visualize

```python
X_scaled.shape
```

```python
#2D
from sklearn.decomposition import PCA
pca = PCA(n_components=3)

X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
#plot
fig= plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:,0], X_pca[:, 1], X_pca[:, 2])

ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
ax.set_title("3d projection")
```

# Analyze the K value

### 1.Elbow method

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit_predict(X_pca)
    wcss.append(kmeans.inertia_)
    
```

```python
knee = KneeLocator(range(1, 11), wcss, curve="convex", direction="decreasing")
optimal_k = knee.elbow
```

```python
print("Best k = ",optimal_k)
```

```python
#plot 
plt.plot(range(1,11), wcss, marker='o') 
plt.xlabel("K")
plt.ylabel("WCSS")
```

## 2. Silhouette Score

```python
from sklearn.metrics import silhouette_score

scores=[]
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, labels)
    scores.append(score)
    
#plot 
plt.plot(range(2,11), scores, marker='o') 
plt.xlabel("K")
plt.ylabel("Silhouette Score")
```

```python
#combined plot

k_range = range(2, 11)
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(k_range, wcss[:len(k_range)], marker='o', color="blue")
ax1.set_xlabel("K") 
ax1.set_ylabel("WCSS") 

ax2 = ax1.twinx() #twinx(): for same x-axis
ax2.plot(k_range, scores[:len(k_range)], marker='x', color="red", linestyle="--")
ax2.set_ylabel("Silhoutte Scores") 
```

## Clustering

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels_kmeans= kmeans.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))

ax = fig.add_subplot(111, projection="3d")

ax.scatter(X_pca[:,0], X_pca[:, 1], X_pca[:, 2], c=labels_kmeans)
```

```python
# Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_clf = AgglomerativeClustering(n_clusters=4, linkage="ward")
labels_agg = agg_clf.fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_pca[:,0], X_pca[:, 1], X_pca[:, 2], c=labels_agg)
```
